1. 개요

- 분석의 목적

요즘 들어 미세먼지 농도가 환경문제로 대조되면서 국민들의 건강이 위협받고 있습니다. 또한 우리 주변에서도 기관지염으로 불편함을 호소하는 친구들을 많이 보면서 미세먼지 문제와 기관지염의 연관성을 발견하게 됐습니다. 이 둘의 연관성을 알 수 있는 데이터를 시각화를 통해 보건계열의 편의성을 도모할 수 있으며 실시간 데이터를 통해 국민들의 편의성을 증진시킬 수 있습니다. 이 데이터들을 이용해 나온 결과를 사용하여 병원에서는 mri 등의 복잡한 절차를 거치지 않고도 기관지 치료를 보다 수월하게 할 수 있고, 해당 날짜의 미세먼지가 심한 지역 국민들은 외출을 자제하거나, 마스크를 착용하는 등의 대비를 할 수 있습니다.

- (공모전의 경우) 문제 설명

공모전 제목 = 환경데이터 활용 및 분석 공모전

미세먼지의 가장 큰 문제점은 국민의 건강입니다. 
숨을 쉬는 과정에서 공기와 함께 미세먼지가 폐 속으로 들어와 폐가 정상적인 기능을 하지 못할 정도로 악영향을 끼칩니다. 또한 면역기능과 폐의 기능을 약화시키고, 이는 천식과 같은 호흡기 질병으로 이어집니다. 이는 인간의 생존수단과 중요한 연관이 있는 농업과 환경도 영향을 끼칩니다. 왜냐하면 미세먼지 농도가 높아지면 광합성을 방해하여 농작물을 생장에도 좋지 않은 영향을 미치기 때문입니다. 
미세먼지로 인해 빗물은 강 산성을 띄게 되고 토양 또한 강 산성으로 변하게 되어서 석회암이나 대리석으로 만들어진 유적들의 부식을 일으킵니다.

- 수집 방법

위에있는 사이트들을 통해 데이터 자료들을 수집하고 그 자료들을 히스토그램이나 산점도와 같은 데이터 시각화 방법을 이용하여 데이터를 분석할것입니다. 

코드 = 판다스(Pandas)예정.

- 수집된 데이터의 구조 및 특성 설명

-서울시 병원 내 호흡기 질병 환자 데이터: 정형 데이터

*특성
정형 데이터

-관계형 데이터 베이스 시스템의 테이블과 같이 고정된 칼럼에 저장되는 데이터와 파일이다.

-지정된 행, 열에 의해 데이터 속성이 구별되는 스프레드 시트 형태의 데이터이다.

-데이터의 스키마 정보를 관리하는 DBMS와 에디터 내용이 저장되는 데이터 저장소로 구분된다.
